```python
data=[45,55,67,45,68,79,98,87,84,82]
s_m=np.mean(data)
s_sd=np.std(data)
print(s_m,s_sd)
n=len(data)
t=t.ppf(0.99,n-1)
L=s_m - (t*(s_sd/np.sqrt(n-1)))
U=s_m + (t*(s_sd/np.sqrt(n-1)))
CI=(L,U)
print("98% confidence interval for population mean=",CI)
```
