```python
import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
```

**image.png**

```python
y=np.array([70,65,90,95,110,115,120,140,155,150])
x=np.array([80,100,120,140,160,180,200,220,240,260])
plt.scatter(x,y)
```

```python
m_x=np.mean(x)
m_y=np.mean(y)
n=len(x)
print(m_x,m_y,n)
```

```python
cov=(np.sum(x*y)-n*m_x*m_y)/n
print(cov)
```

```python
sx= np.sqrt((np.sum(x*x)-n*m_x*m_x)/n)
sy= np.sqrt((np.sum(y*y)-n*m_y*m_y)/n)
r=cov/(sx*sy)
print(r)
```

* Ho: coefficent of correlation =0
* Ha: !=0

```python
t=r*np.sqrt(n-2)/(np.sqrt(1-r**2))
print("t-statistics=",t)
t_cri=st.t.ppf(0.975,n-2)
print("t-critical: ",t_cri)
p=(st.t.sf(t,n-2))*2
print("p-value= ",p)
```

* Reject Null hypothesis

```python
corr,pvalue=st.pearsonr(x,y)
print(corr,pvalue)
```

```python
x=np.array([80,100,120,140,160,180,200,220,240,260])
y=np.array([70,65,90,95,110,115,120,140,155,150])
plt.scatter(x,y)
```

```python
n=len(x)
m_x=np.mean(x)
m_y=np.mean(y)
sxy=np.sum(x*y)-n*m_x*m_y
sx2=np.sum(x*x)-n*m_x*m_x
b_1=sxy/sx2
b_0=m_y-b_1*m_x
print("Coeffiecient beta_0 and beta_1= ",b_0,b_1)
plt.plot(x,y)
```

```python
plt.scatter(x,y,color="m",marker="o",s=30)
y_pred=b_0+b_1*x
plt.plot(x,y_pred,color="g")
```

**Hypothesis test for B1**
* Ho:B1=0
* Ha:B1!=0

```python
#  y_pred=b_0+(b_1)*x
num=np.sum((y-y_pred)**2)
den=(np.sum((x-m_x)**2))*(n-2)
se_b1=np.sqrt(num/den)
tstats=b_1/se_b1
print("t-statistics= ",tstats)
print("t-critical= ",st.t.ppf(0.975,n-2))
```

**Hypothesis test for B0**
* Ho:B0=0

```python
# y_pred=b_0+(b_1)*x
ft=(np.sum((y-y_pred)**2))/(n-2)
st=(1/n)+((m_x**2)/(np.sum((x-m_x)**2)))
se_b0=np.sqrt(ft*st)
tstats=b_0/se_b0
print("t-statistics= ",tstats)
import scipy.stats as st
print("t-critical=",st.t.ppf(0.975,n-2))
```

**Reject Null Hypothesis**

```python
exp_var=np.sum((y_pred-m_y)**2)
tot_var=np.sum((y-m_y)**2)
r2=exp_var/tot_var
print("Coefficient of determination= ",r2)
```
