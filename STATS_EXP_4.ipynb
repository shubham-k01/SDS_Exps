```python
l=1.8
p=poisson.pmf(0,l)
print("(i)Probability that computer will function without breakdown ",p)
```

The Erlang distribution is a specific case of the Gamma distribution. The difference between Erlang and Gamma is that in a Gamma distribution, k can be a non-integer (positive real number) and in Erlang, k is positive integer only.



```python
from scipy.stats import gamma
```

```python
p= 1- gamma.cdf(12,3,0,2)
print("Probability that power will be inadequate on a given day= ",p)
```
